## Create a function that uses original column unique values to generate the appropriate unique IDs for each section.

````python
def get_unique_ids(df, col, new_col_name): #Where df is the dataframe and col is unique column string
    
    #create a list of unique names
    unique_names = list(df[col].unique())
    
    #create a new column of plaintiff names
    df[new_col_name] = df[col]
    
    #Create a dictionary of unique  with associated randomly generated unique ids
    unique_names_ids = dict(zip(unique_names, np.random.randint(1, 1000, len(unique_names))))
    
    #Replace new column with unique ids using .replace and the unique_names_ids
    df[new_col_name] = df[new_col_name].replace(unique_names_ids)
    
    return df
````

### SQL QUERY 1:
Pull:
- `case_id` from `cases`
- `plaintiff_name` from `plaintiffs`, joining on `plaintiff_id`
- `attorney` from `attornies`, joining on `attorney_id`
- `lien_description` from `liens` joining on `lien_id`
<br>
For liens held by lienholder General Assembly, which is associated with the lienholder_id of 794, which can be referenced through `liens.lienholder_id = 794`.
<br>
<br>
***
SQL Code:

```
SELECT cases.case_id, 
plaintiffs.plaintiff_name, 
attornies.attorney, 
liens.lien_description 
FROM cases 
JOIN plaintiffs 
ON cases.plaintiff_id = plaintiffs.plaintiff_id 
JOIN attornies 
ON cases.attorney_id = attornies.attorney_id 
JOIN liens 
ON cases.case_id=liens.case_id 
WHERE liens.lienholder_id = 794;
```

### SQL Query 2:
Pull all unique attornies with their contact info(phone and email) that are assigned to one or more cases that haven't had a case update in 72 days. In this particular dataset, we can see that there is only 1 case that has been updated in the past 72 days and it is the only value from 2019.
<br>
***
SQL Code:
<br>
```
SELECT attornies.attorney, 
attornies.attorney_phone, 
attornies.attorney_email 
FROM attornies 
JOIN cases 
ON attornies.attorney_id = cases.attorney_id 
JOIN case_updates 
ON cases.case_id = case_updates.case_id 
WHERE case_updates.agreement_date 
NOT LIKE '%2019' 
GROUP BY cases.attorney_id 
HAVING COUNT(*)>=1; 
```
<br><br>
Output:
```
[('Julian Michaels', '917-555-1234', 'jm@fake.gmail.com'),
 ('Lisle Von Trapp', '678-555-1234', 'lvt@fake.gmail.com'),
 ('Joseph Breitstein', '718-555-1234', 'jb@fake.gmail.com'),
 (None, None, None)]
 ```

````python
engine.execute("SELECT attornies.attorney, attornies.attorney_phone, attornies.attorney_email FROM attornies JOIN cases ON attornies.attorney_id = cases.attorney_id JOIN case_updates ON cases.case_id = case_updates.case_id WHERE case_updates.agreement_date NOT LIKE '%2019' GROUP BY cases.attorney_id HAVING COUNT(*)>=1 ").fetchall()
````

### SQL Query 3:
Rank Law Firms(id, name) by number of cases associate with it. 
<br>
We can find the number of cases associated with each law firm by joining the `law_firm` table with the `party_relationships` table as they share the `lawfirm_id` column. From there, I joined the `party_relationships` table with the `cases` table because they share the `attorney_id` column and because the `cases` table allows us to discover how many cases are associate with some grouping factor, in this case, `lawfirm_id`. I then ordered by the count in descending order.
***
SQL CODE
```
SELECT law_firms.lawfirm_id, 
law_firms.lawfirm 
FROM law_firms 
JOIN party_relationships 
ON law_firms.lawfirm_id = party_relationships.lawfirm_id 
JOIN cases 
ON party_relationships.attorney_id=cases.attorney_id 
GROUP BY law_firms.lawfirm_id 
ORDER BY COUNT(*) DESC;
```
<br><br>

Output:
```
[(1, 'Fake NYC Attornies'), (2, 'Fake Brooklyn Attornies')]
```

````python
engine.execute("SELECT law_firms.lawfirm_id, law_firms.lawfirm FROM law_firms JOIN party_relationships ON law_firms.lawfirm_id = party_relationships.lawfirm_id JOIN cases ON party_relationships.attorney_id=cases.attorney_id GROUP BY law_firms.lawfirm_id ORDER BY COUNT(*) DESC").fetchall()
````
